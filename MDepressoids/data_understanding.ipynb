```python
def string_to_bd(string):
    pattern1=u"\D+"
    pattern2=u"FMRI"
    pattern3=u"EEG"
    pattern4=u"_\S+\_"
    r=bd_fmri.iloc[30,0]
    st1= re.findall(pattern1,string)
    if len(st1)!=0:
        st1= st1[0]
    else:
        st1=0
    st2= re.findall(pattern2,string)#.lower()
    if len(st2)!=0:
        st2=st2[0].lower()
    else:
        st2=0
    
    st3= re.findall(pattern3,string)#.lower()
    if len(st3)!=0:
        st3=st3[0].lower()
    else:
        st3=0
    st4= re.findall(pattern4,string)
    if len(st4)!=0:
        st4=st4[0].lower()
        st4= st4.split("_" )[1]
        st4=st4.lower()
    
    else:
        st4=0
    
    
    return st2,st3,st4
r=bd_fmri.iloc[13,0]
fm=[]
ega=[]
name=[]
for i in range(bd_fmri.shape[0]):
    r=bd_fmri.iloc[i,0]
    #print string_to_bd(r)
    f,e,n=string_to_bd(r)
    fm.append(f)
    ega.append(e)
    name.append(n)
Datas={"fmri":fm,"eeg":ega,"name":name}    
bd_fmri_ok=pd.DataFrame(Datas)
bd_fmri_ok.head()
```
